**2 Dimensi 1 grup dengan metode jacobi**

Persamaan difusi 1 grup
$$ -\vec{\nabla} D \vec{\nabla} \phi +  \sum_{r} \phi = \frac{1}{K_eff} v \sum_{f} \phi $$
<br> Karena 2 dimensi maka diintegralkan ke arah i, j
$$ -\int_{i,j}\vec{\nabla}D \vec{\nabla}\phi d^3\vec r + \int_{i,j} \sum_{r} \phi d^3\vec r= \int_{i,j} \frac{1}{K_eff} v \sum_{f} \phi d^3\vec r $$
<br> Dengan teorema divergensi
$$ -\int_{i,j} D \vec{\nabla}\phi \vec{dA} + \sum_{r}^{i,j} \phi^{i,j} V^{i,j}= \frac{1}{K_eff} v \sum_{f}^{i,j} \phi^{i,j} V^{i,j} $$
<br> Pada suku difusi
$$ -\int_{i+1,j} D \vec{\nabla}\phi \vec{dA} + \int_{i-1,j} D \vec{\nabla}\phi \vec{dA} - \int_{i,j+1} D \vec{\nabla}\phi \vec{dA}+ \int_{i,j-1} D \vec{\nabla}\phi\vec{dA}$$
$$ -D^{{i+1,j}} \frac{\phi^{i+1,j}-\phi^{i,j}} {r_{i+1} -r_{i}} A^{i+1,j}+D^{{i-1,j}} \frac{\phi^{i,j} - \phi^{i-1,j}} {r_{i} -r_{i-1}} A^{i-1,j} -D^{{i,j+1}} \frac{\phi^{i,j+1}-\phi^{i,j}} {z_{j+1} -z_{j}} A^{i,j+1} + D^{{i,j-1}} \frac{\phi^{i,j} - \phi^{i,j-1}} {z_{j} -z_{j-1}} A^{i,j-1}$$
<br> Maka persamaan menjadi
$$ -D^{{i+1,j}} \frac{\phi^{i+1,j}-\phi^{i,j}} {r_{i+1} -r_{i}} A^{i+1,j}+D^{{i-1,j}} \frac{\phi^{i,j} - \phi^{i-1,j}} {r_{i} -r_{i-1}} A^{i-1,j} -D^{{i,j+1}} \frac{\phi^{i,j+1}-\phi^{i,j}} {z_{j+1} -z_{j}} A^{i,j+1} + D^{{i,j-1}} \frac{\phi^{i,j} - \phi^{i,j-1}} {z_{j} -z_{j-1}} A^{i,j-1} $$
$$ + \sum_{r}^{i,j} \phi^{i,j} V^{i,j} = \frac{1}{K_eff} v \sum_{f}^{i,j} \phi^{i,j} V^{i,j} $$

Kelompokkan $\phi$
<br> $$ \phi^{i,j-1} (\frac{-D^{i,j-1}}{z_{j} -z_{j-1}} A^{i,j-1}) + \phi^{i-1,j} (\frac{-D^{i-1,j}}{r_{i} -r_{i-1}} A^{i-1,j}) $$
$$ + \phi^{i,j} (\frac{D^{i+1,j}}{r_{i+1} -r_{i}} A^{i+1,j} + \frac{D^{i-1,j}}{r_{i} -r_{i-1}} A^{i-1,j}
 + \frac{D^{i,j+1}}{z_{j+1} -z_{j}} A^{i,j+1} + \frac{D^{i,j-1}}{z_{j} -z_{j-1}} A^{i,j-1} + \sum_{r}^{i,j} V^{i,j}) $$
$$ + \phi^{i+1,j} (\frac{-D^{i+1,j}}{r_{i+1} -r_{i}} A^{i+1,j}) + \phi^{i,j+1} (\frac{-D^{i,j+1}}{z_{j+1} -z_{j}} A^{i,j+1}) = \frac{1}{K_eff} v \sum_{f}^{i,j} \phi^{i,j} V^{i,j} $$
<br> Misalkan
$$ \gamma^{i,j} = \frac{D^{i,j-1}}{z_{j} -z_{j-1}} A^{i,j-1}.......(1)$$
 $$ \alpha^{i,j} = \frac{D^{i-1,j}}{r_{i} -r_{i-1}} A^{i-1,j}......(2)$$
 $$ \beta^{i,j} = \frac{D^{i+1,j}}{r_{i+1} -r_{i}} A^{i+1,j} + \frac{D^{i-1,j}}{r_{i} -r_{i-1}} A^{i-1,j} + \frac{D^{i,j+1}}{z_{j+1} -z_{j}} A^{i,j+1} + \frac{D^{i,j-1}}{z_{j} -z_{j-1}} A^{i,j-1} + \sum_{r}^{i,j} V^{i,j}.......(3)
 $$
 $$ S^{i,j} = \frac{1}{K_eff} v \sum_{f}^{i,j} \phi^{i,j} V^{i,j}......(4) $$

Maka persamaan menjadi
$$ -\gamma^{i,j} \phi^{i, j-1} -\alpha^{i,j} \phi^{i-1,j} + \beta^{i,j} \phi^{i,j} - \alpha^{i+1,j} \phi^{i+1,j} - \gamma^{i,j+1} \phi^{i,j+1} = S^{i,j} $$
Kemudian di buat kedalam matriks
$$ A \phi = S $$

**Algoritma**
1. Tebak semua $\phi$ dan $K_{eff}$
2. Hitung sumber $S^{i,j}$
3. Menggunakan metode iteratif jacobi
$$ \phi^{i,j} = \frac{S^{i,j}+\gamma^{i,j-1}\phi^{i,j-1}+\alpha^{i-1,j}\phi+\alpha^{i+1,j}\phi+\gamma^{i,j+1}\phi}{\beta^{i,j}} $$
4. Hitung total fisi
$$ f_{n+1} = \sum_g \sum_j \sum_i v \sum_{f}^{i,j} \phi_{n+1}^{i,j} V^{i,j} $$
5. Update nilai $K_{eff}$
$$ K_{eff}^{n+1} = K_{eff}^{n} \frac{F_{n+1}} {F_n} $$
6. Cek konvergensi
$$ \lvert {\frac{K_{eff}^{n+1}-K_{eff}^{n}}{K_{eff}^{n+1}+K_{eff}^{n}}} \rvert < \epsilon $$
7. Bila belum konvergen kembali ke langkah 2

### **Penerapan algoritma**

Initial value

```python
# Import library
import numpy as np
import math

# Initial Constants
sig_s = 0.1
sig_a = 0.01
sig_r = sig_s + sig_a
nu_sigf = 0.1
keff_awal = 1
fluks_tebakan = 2200
d = 1
r = np.linspace(0, 1, 10)
z = np.linspace(0, 9, 10)
```

Membuat matriks

```python
# Membuat matriks gamma dengan persamaan (1)
gamma = []
for i in range(len(r)):
    for j in range(len(z)):
        g = -(d/(z[j]-z[j-1]))*(2*(z[j]-z[j-1]))
        gamma.append(g)

# Membuat matriks alpha dengan persamaan (2)
alpha = []
for j in range(len(z)):
    for i in range(len(r)):
        a = -(d/(r[i]-r[i-1]))*(math.pi*(r[i]-r[i-1])**2)
        alpha.append(a)
alpha[0] = alpha[-1]

# Membuat matriks beta dengan persamaan (3)
def get_beta (r, z, sig_r):
    beta1 = []
    beta2 = []
    # sisi r
    for j in range(len(z)):
        for i in range(len(r)):
            if i < len(r) - 1:  # Menghindari indeks keluar
                suku1 = (d/(r[i+1]-r[i]))*(math.pi*(r[i+1]*r[i-1])**2)
            else:
                suku1 = (d/(r[1]))*(math.pi*(r[1])**2)
            if i > 0:
                suku2 = (d/(r[i]-r[i-1]))*(math.pi*(r[i]-r[i-1])**2)
            else:
                suku2 = (d/(r[1]))*(math.pi*(r[1])**2)
            jum = suku1 + suku2
            beta1.append(jum)
    # sisi z
    for i in range(len(r)):
        for j in range(len(z)):
            if j < len(z) - 1:  # Menghindari indeks keluar
                suku3 = (d / (z[j+1] - z[j])) *(2*(z[j+1]-z[j]))
            else:
                suku3 = (d/(z[1]))*(2*(z[1]))
            if j > 0:
                suku4 = (d/(z[j]-z[j-1]))*(2*(z[j]-z[j-1]))
            else:
                suku4 = (d/(z[1]))*(2*(z[1]))
            jum = suku3 + suku4
            beta2.append(jum)
    beta = np.array(beta1) + np.array(beta2) + (sig_r*2*math.pi*r[-1]*(r[-1]+z[-1]))
    return beta
beta = get_beta(r,z, sig_r)
# print(beta)

# Membuat matriks fluks dengan nilai tebakkan
phi = np.ones((len(beta), 1))

# Dengan menggunakan persamaan (4) dibuat matriks sumber
s = (1/keff_awal)*phi
s = np.array(s)

# Membuat matriks A dengan nilai nol
A = np.zeros((len(beta), len(beta)))
# Baris pertama (beta di atas, alpha di sebelahnya)
A[0, 0] = beta[0]
A[0, 1] = alpha[0]
A[0, 50] = gamma[0] # Gamma di sisi atas
# Baris berikutnya (beta diagonal dengan alpha di sub diagonal dan gamma di atas dan bawah)
for i in range(1, len(beta)-1):
    A[i, i] = beta[i]  # Diagonal utama
    A[i, i - 1] = alpha[i]  # Sub-diagonal
    A[i, i + 1] = alpha[i]  # Sub-diagonal
    # Gamma di atas diagonal untuk baris atas
    if i < 50:  
        A[i, i + 50] = gamma[i]
    # Gamma di bawah diagonal untuk baris bawah
    if i >= 50:  
        A[i, i - 50] = gamma[i]
# Baris terakhir
A[-1, -1] = beta[-1]  # Beta di diagonal utama terakhir
A[-1, -2] = alpha[-1]  # Alpha di bawah diagonal utama
A[99, 49] = gamma[-1]  # Gamma di sisi bawah
# # Cek bentuk matriks A
# import pandas as pd
# df = pd.DataFrame(A)
# df.to_csv("matriksA.csv", index = False)
```

Perhitungan fluks dengan metode jacobi

```python
# Hitung total fisi awal
fisi_awal = len(A) * fluks_tebakan * nu_sigf * math.pi*r[-1]**2*z[-1]

# Inisialisasi fluks dan parameter lainnya
fluks_n = np.zeros_like(s) # Membuat matriks fluks baru berdimensi sama dengan sumber
fisi = [fisi_awal]
keff = [keff_awal]
all_fluks = [] # Untuk menaruh seluruh perhitungan fluks

# Fungsi Jacobi
def jacobi(new_fluks, matriks_A, matriks_sumber, fisi, keff, toleransi, nusigf, max_iter):
    iter_count = 0
    while iter_count < max_iter:
        iter_count += 1
        # Hitung nilai fluks baru
        for i in range(len(matriks_A)):
            s1 = np.dot(matriks_A[i, :i], new_fluks[:i])
            s2 = np.dot(matriks_A[i, i + 1:], new_fluks[i + 1:])
            new_fluks[i] = (matriks_sumber[i] - s1 - s2) / matriks_A[i, i]
            #print(new_fluks)
        # Simpan hasil fluks baru
        all_fluks.append(new_fluks.copy())
        # Hitung fisi baru
        f_i = np.sum(new_fluks) * nusigf * math.pi*r[-1]**2*z[-1]
        #(f_i)
        fisi.append(f_i)
        # Hitung keff baru
        keff_i = keff[-1] * (fisi[-1] / fisi[-2])
        keff.append(keff_i)
        # Cek keff
        cek_keff = abs((keff[-2] - keff[-1]) / (keff[-2] + keff[-1]))
        #print("cek keff: ", cek_keff)
        # Cek konvergensi
        if cek_keff <= toleransi:
            print("Selesai")
            return all_fluks
        # Update sumber dengan nilai fluks baru
        matriks_sumber = (1 / keff[-1]) * new_fluks
    return all_fluks

# Panggil Fungsi Jacobi
all_fluks = jacobi(fluks_n, A, s, fisi, keff, toleransi=1.0e-3, nusigf=nu_sigf, max_iter=400)
print(len(all_fluks))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


fluks = all_fluks[0]
# Reshape flux data into a 2D array to fit r and z
fluks = fluks.reshape(len(r), len(z))

# Create a meshgrid for r and z
R, Z = np.meshgrid(r, z, indexing='ij')

# Plotting
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(Z, R, fluks, cmap='viridis', edgecolor='k')

# Adding labels
ax.set_title("Distribusi Fluks Neutron ke Arah Radial dan z")
ax.set_xlabel("mesh ke arah r")
ax.set_ylabel("mesh ke arah z")
ax.set_zlabel("Fluks neutron")
ax.view_init(elev=45, azim=60, roll=15)
plt.show()
```
