Persamaan difusi 1 grup
$$ -\vec{\nabla} D \vec{\nabla} \phi +  \sum_{r} \phi = \frac{1}{K_eff} v \sum_{f} \phi $$
<br> Karena 2 dimensi maka diintegralkan ke arah i, j
$$ -\int_{i,j}\vec{\nabla}D \vec{\nabla}\phi d^3\vec r + \int_{i,j} \sum_{r} \phi d^3\vec r= \int_{i,j} \frac{1}{K_eff} v \sum_{f} \phi d^3\vec r $$
<br> Dengan teorema divergensi
$$ -\int_{i,j} D \vec{\nabla}\phi \vec{dA} + \sum_{r}^{i,j} \phi^{i,j} V^{i,j}= \frac{1}{K_eff} v \sum_{f}^{i,j} \phi^{i,j} V^{i,j} $$
<br> Pada suku difusi
$$ -\int_{i+\frac{1}{2},j} D \vec{\nabla}\phi \vec{dA} + \int_{i-\frac{1}{2},j} D \vec{\nabla}\phi - \int_{i,j+\frac{1}{2}} D \vec{\nabla}\phi + \int_{i,j-\frac{1}{2}} D \vec{\nabla}\phi$$
$$ -D^{{i+\frac{1}{2},j}} \frac{\phi^{i+1,j}-\phi^{i,j}} {r_{i+1} -r_{i}} A^{i+\frac{1}{2},j}+D^{{i-\frac{1}{2},j}} \frac{\phi^{i,j} - \phi^{i-1,j}} {r_{i} -r_{i-1}} A^{i-\frac{1}{2},j} -D^{{i,j+\frac{1}{2}}} \frac{\phi^{i,j+1}-\phi^{i,j}} {z_{j+1} -z_{j}} A^{i,j+\frac{1}{2}} + D^{{i,j-\frac{1}{2}}} \frac{\phi^{i,j} - \phi^{i,j-1}} {z_{j} -z_{j-1}} A^{i,j-\frac{1}{2}}$$
<br> Maka persamaan menjadi
$$ -D^{{i+\frac{1}{2},j}} \frac{\phi^{i+1,j}-\phi^{i,j}} {r_{i+1} -r_{i}} A^{i+\frac{1}{2},j}+D^{{i-\frac{1}{2},j}} \frac{\phi^{i,j} - \phi^{i-1,j}} {r_{i} -r_{i-1}} A^{i-\frac{1}{2},j} -D^{{i,j+\frac{1}{2}}} \frac{\phi^{i,j+1}-\phi^{i,j}} {z_{j+1} -z_{j}} A^{i,j+\frac{1}{2}} + D^{{i,j-\frac{1}{2}}} \frac{\phi^{i,j} - \phi^{i,j-1}} {z_{j} -z_{j-1}} A^{i,j-\frac{1}{2}} + \sum_{r}^{i,j} \phi^{i,j} V^{i,j} = \frac{1}{K_eff} v \sum_{f}^{i,j} \phi^{i,j} V^{i,j} $$

Kelompokkan $\phi$
<br> $$ \phi^{i,j-1} (\frac{-D^{i,j-\frac{1}{2}}}{z_{j} -z_{j-1}} A^{i,j-\frac{1}{2}}) + \phi^{i-1,j} (\frac{-D^{i-\frac{1}{2},j}}{r_{i} -r_{i-1}} A^{i-\frac{1}{2},j}) + \phi^{i,j} (\frac{D^{i+\frac{1}{2},j}}{r_{i+1} -r_{i}} A^{i+\frac{1}{2},j} + \frac{D^{i-\frac{1}{2},j}}{r_{i} -r_{i-1}} A^{i-\frac{1}{2},j} + \frac{D^{i,j+\frac{1}{2}}}{z_{j+1} -z_{j}} A^{i,j+\frac{1}{2}} + \frac{D^{i,j-\frac{1}{2}}}{z_{j} -z_{j-1}} A^{i,j-\frac{1}{2}} + \sum_{r}^{i,j} V^{i,j}) $$
$$ + \phi^{i+1,j} (\frac{-D^{i+\frac{1}{2},j}}{r_{i+1} -r_{i}} A^{i+\frac{1}{2},j}) + \phi^{i,j+1} (\frac{-D^{i,j+\frac{1}{2}}}{z_{j+1} -z_{j}} A^{i,j+\frac{1}{2}}) = \frac{1}{K_eff} v \sum_{f}^{i,j} \phi^{i,j} V^{i,j} $$
<br> Misalkan
$$ \gamma^{i,j} = \frac{D^{i,j-\frac{1}{2}}}{z_{j} -z_{j-1}} A^{i,j-\frac{1}{2}}.......(1)$$
 $$ \alpha^{i,j} = \frac{D^{i-\frac{1}{2},j}}{r_{i} -r_{i-1}} A^{i-\frac{1}{2},j}......(2)$$
 $$ \beta^{i,j} = \frac{D^{i+\frac{1}{2},j}}{r_{i+1} -r_{i}} A^{i+\frac{1}{2},j} + \frac{D^{i-\frac{1}{2},j}}{r_{i} -r_{i-1}} A^{i-\frac{1}{2},j} + \frac{D^{i,j+\frac{1}{2}}}{z_{j+1} -z_{j}} A^{i,j+\frac{1}{2}} + \frac{D^{i,j-\frac{1}{2}}}{z_{j} -z_{j-1}} A^{i,j-\frac{1}{2}} + \sum_{r}^{i,j} V^{i,j}.......(3)
 $$
 $$ S^{i,j} = \frac{1}{K_eff} v \sum_{f}^{i,j} \phi^{i,j} V^{i,j}......(4) $$

Maka persamaan menjadi
$$ -\gamma^{i,j} \phi^{i, j-1} -\alpha^{i,j} \phi^{i-1,j} + \beta^{i,j} \phi^{i,j} - \alpha^{i+1,j} \phi^{i+1,j} - \gamma^{i,j+1} \phi^{i,j+1} = S^{i,j} $$
Kemudian di buat kedalam matriks
$$ A \phi = S $$

**Algoritma**
1. Tebak semua $\phi$ dan $K_{eff}$
2. Hitung sumber $S^{i,j}$
3. Menggunakan metode iteratif jacobi
$$ \phi^{i,j} = \frac{S^{i,j}+\gamma^{i,j-1}\phi^{i,j-1}+\alpha^{i-1,j}\phi+\alpha^{i+1,j}\phi+\gamma^{i,j+1}\phi}{\beta^{i,j}} $$
4. Hitung total fisi
$$ f_{n+1} = \sum_g \sum_j \sum_i v \sum_{f}^{i,j} \phi_{n+1}^{i,j} V^{i,j} $$
5. Update nilai $K_{eff}$
$$ K_{eff}^{n+1} = K_{eff}^{n} \frac{F_{n+1}} {F_n} $$
6. Cek konvergensi
$$ \lvert {\frac{K_{eff}^{n+1}-K_{eff}^{n}}{K_{eff}^{n+1}+K_{eff}^{n}}} \rvert < \epsilon $$
7. Bila belum konvergen kembali ke langkah 2

### **Penerapan algoritma**

```python
# Import library
import numpy as np
import matplotlib.pyplot as plt

# Initial Constants
sig_s = 0.1
sig_a = 0.01
sig_r = sig_s + sig_a
d = 1
nu_sigf = 0.1
nilai_i = 30
nilai_j = 30
delta_z = 5
delta_r = 5
keff_awal = 1
fluks_tebakan = 2200

# Dengan menggunakan persamaan (1),(2) dan (3) untuk menghitung alpha, beta, dan gamma
alpha = -d / delta_r
beta = 2 * (d / delta_r) + 2 * (d / delta_z) + sig_r
gamma = -d / delta_r

# Membuat matriks fluks dengan nilai tebakkan
phi = np.ones((N+2, 1))

# Dengan menggunakan persamaan (4) dibuat matriks sumber
s = (1/keff_awal)*phi
s = np.array(s)

# Ukuran matriks dan konstanta
N = 60  # Ukuran grid
# Membuat matriks A dengan zeros
A = np.zeros((N + 2, N + 2))
# Baris pertama (gamma di atas, alpha di sebelahnya)
A[0, 0] = beta
A[0, 1] = alpha
A[0, 30] = gamma  # Gamma di sisi atas
# Baris berikutnya (beta diagonal dengan alpha di sub diagonal dan gamma di atas dan bawah)
for i in range(1, N + 1):
    A[i, i] = beta  # Diagonal utama
    A[i, i - 1] = alpha  # Sub-diagonal
    A[i, i + 1] = alpha  # Sub-diagonal
    # Gamma di atas diagonal untuk baris atas
    if i < 31:  
        A[i, i + 30] = gamma
    # Gamma di bawah diagonal untuk baris bawah
    if i >= 31:  
        A[i, i - 32] = gamma
# Baris terakhir (gamma di bawah, alpha di sebelahnya)
A[N + 1, N] = alpha  # Alpha di bawah diagonal utama
A[N + 1, N + 1] = beta  # Beta di diagonal utama terakhir
A[N + 1, N - 31] = gamma  # Gamma di sisi bawah

# Hitung total fisi awal
fisi_awal = 60 * fluks_tebakan * nu_sigf

# Inisialisasi fluks dan parameter lainnya
x = np.zeros_like(s)
fluks_n = np.zeros_like(x)
fisi = [fisi_awal]
keff = [keff_awal]
all_fluks = []

# Fungsi Jacobi
def jacobi(new_fluks, matriks_A, matriks_sumber, fisi, keff, toleransi, nu_sigf, max_iter=1000):
    iter_count = 0
    while iter_count < max_iter:
        iter_count += 1
        # Hitung nilai fluks baru
        for i in range(matriks_A.shape[0]):
            s1 = np.dot(matriks_A[i, :i], new_fluks[:i])
            s2 = np.dot(matriks_A[i, i + 1:], new_fluks[i + 1:])
            new_fluks[i] = (matriks_sumber[i] - s1 - s2) / matriks_A[i, i]
        # Simpan hasil fluks baru
        all_fluks.append(new_fluks.copy())
        # Hitung fisi baru
        f_i = np.sum(new_fluks) * nu_sigf
        fisi.append(f_i)
        # Hitung keff baru
        keff_i = keff[-1] * (fisi[-1] / fisi[-2])
        keff.append(keff_i)
        # Cek keff
        cek_keff = abs((keff[-2] - keff[-1]) / (keff[-2] + keff[-1]))
        # Cek konvergensi
        if cek_keff <= toleransi:
            break
        # Update sumber dengan nilai fluks baru
        matriks_sumber = (1 / keff[-1]) * new_fluks
    return fisi, keff, all_fluks

# Panggil Fungsi Jacobi
hasil_fisi, hasil_keff, all_fluks = jacobi(fluks_n, A, s, fisi, keff, toleransi=1e-5, nu_sigf=nu_sigf)
```
